```python
import networkx as nx
from itertools import combinations
import matplotlib.pyplot as plt
```

```python
G=nx.karate_club_graph()
pos=nx.spring_layout(G)
colors = []
for i in range(G.number_of_nodes()):
    colors.append(nx.algorithms.core.core_number(G)[i])
nx.draw(G, pos, with_labels=True, node_size=200)
nx.draw_networkx_nodes(G, pos, nodelist=G.nodes(), node_color=colors, node_size=200, label=True)
```

```python
G1=nx.les_miserables_graph()
pos=nx.spring_layout(G1)
colors = []
for i in G1.nodes():
    colors.append(nx.algorithms.core.core_number(G1)[i])
nx.draw(G1, pos, with_labels=False, node_size=100)
nx.draw_networkx_nodes(G1, pos, nodelist=G1.nodes(), node_color=colors, node_size=100, label=False)
```
