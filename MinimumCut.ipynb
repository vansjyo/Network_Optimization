```python
# Setting up environment
import networkx as nx
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [5, 5]
```

```python
# Define the network
num_nodes=8
G=nx.DiGraph()
G.add_nodes_from(range(1,num_nodes+1))
G.add_weighted_edges_from([(1,2,8),(1,3,7),(1,4,4),
                           (2,3,2),(2,5,3),(2,6,9),
                           (3,6,6),(3,4,5),
                           (4,6,1),(4,7,2),
                           (6,5,3),(6,8,5),(6,7,4),
                           (5,8,9),(7,8,8)]) 
print("Nodes: \n", G.nodes(data=True))
print("Edges: \n", G.edges(data=True))
```

```python
# Draw the Graph
pos=nx.shell_layout(G)
edge_labels = nx.get_edge_attributes(G,'weight')
nx.draw(G,pos, node_size=500, width=0.5, with_labels=True)
nx.draw_networkx_edge_labels(G,pos, edge_labels=edge_labels)
nx.draw_networkx_nodes(G,pos, nodelist=[i for (i,d) in G.nodes(data=True)], node_size=500) 
```

```python
# Find the minimum s-t cut
cut_value, partition = nx.minimum_cut(G, 1, 8, capacity='weight')
reachable, non_reachable = partition
```

```python
# Print Answer
print("S:", reachable)
print("S':", non_reachable)
print("Min cut value / Max Flow: ", cut_value, "units")
```
